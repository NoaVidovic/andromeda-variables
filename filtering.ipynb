```python
import numpy as np
import pandas as pd

from astroquery.gaia import Gaia
from zero_point import zpt
```

```python
query = """
SELECT source_id, ra, dec, parallax, parallax_error, pmra, pmra_error, pmdec, pmdec_error,
       astrometric_params_solved, pseudocolour, pseudocolour_error, phot_g_mean_mag,
       phot_bp_mean_mag, phot_rp_mean_mag, bp_rp, radial_velocity, radial_velocity_error,
       phot_variable_flag, ecl_lat, nu_eff_used_in_astrometry, phot_g_n_obs
FROM gaiadr3.gaia_source
WHERE in_andromeda_survey = 't'
"""

job = Gaia.launch_job_async(query)

r = job.get_results().to_pandas()
r.to_csv('jobresult00_unfiltered.csv', index=False)
r
```

```python
query = """
SELECT source_id, ra, dec, parallax, parallax_error, pmra, pmra_error, pmdec, pmdec_error,
       astrometric_params_solved, pseudocolour, pseudocolour_error, phot_g_mean_mag,
       phot_bp_mean_mag, phot_rp_mean_mag, bp_rp, radial_velocity, radial_velocity_error,
       phot_variable_flag, ecl_lat, nu_eff_used_in_astrometry, phot_g_n_obs
FROM gaiadr3.gaia_source
WHERE in_andromeda_survey = 't'
    AND astrometric_params_solved IN (31, 95)
    AND bp_rp BETWEEN -1 AND 4
    AND phot_g_mean_mag > 16
    AND abs(pmra) < 0.2 + 2*pmra_error
    AND abs(pmdec) < 0.2 + 2*pmdec_error
"""

job = Gaia.launch_job_async(query)

r = job.get_results().to_pandas()
r.to_csv('jobresult01_pm_phot_filter.csv', index=False)
r
```

```python
zpt.load_tables()
parallax_corrections = r.apply(zpt.zpt_wrapper, axis=1)

r['parallax'] += parallax_corrections
r.to_csv('jobresult02_corrected_parallax.csv', index=False)
r
```

```python
andromeda_parallax = 1/800  # mas
r = r.loc[(r['parallax'] - 2*r['parallax_error'] < andromeda_parallax) & (andromeda_parallax < r['parallax'] + 2*r['parallax_error'])]
r.to_csv('jobresult03_parallax_filter.csv', index=False)
r
```
